```python
import json
import os
import csv
```

```python
id_to_text={}
```

```python
for filename in os.listdir("/home1/zishan/raghav/Emotion-Prediction/TwitterData/"):
    with open(f'/home1/zishan/raghav/Emotion-Prediction/TwitterData/{filename}') as f:
        data = json.load(f)
        for line in data:
            tweet_id = line['id']
            tweet_text = line['text']
            id_to_text[tweet_id]=tweet_text
#             print(tweet_id, tweet_text,end='\n******\n')
```

```python
id_to_emotion={}
```

```python
with open("/home1/zishan/raghav/Emotion-Prediction/id_annotated.tsv") as f:
    rd = csv.reader(f, delimiter="\t")
    for line in rd:
        tweet_id = line[0]
        tweet_emotion, tweet_language = line[1].split(',')
        print(tweet_language)
        if tweet_language =='Hindi':
            id_to_emotion[tweet_id]= tweet_emotion
#         print(tweet_emotion)
#         print(tweet_language)
```

```python
for tweet_id, tweet_emotion in id_to_emotion.items():
    try:
        tweet_text = id_to_text[tweet_id]
        print(tweet_text, tweet_emotion)
    except Exception as e:
        continue
```
